## Importing libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline

ir = 'C:/Users/91932/Desktop/technohacks/IRIS.csv'
iris=pd.read_csv(ir)
iris.info()
iris[0: ]
```

```python
g = sns.pairplot(iris, hue='species', markers='+', vars=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = iris.iloc[:, -1].values

plt.figure(figsize=(8, 6))

for label in np.unique(labels):
    cluster_data = iris.iloc[labels == label, :-1]
    plt.scatter(cluster_data.iloc[:, 0], cluster_data.iloc[:, 1], label=f"Label {label}")

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Data Clustering based on Labels")
plt.legend()
plt.show()
```

```python
n_test_samples = 3

test_data = iris.sample(n=n_test_samples, random_state=42)

train_data = iris.drop(test_data.index)

print("test data:")
print(test_data)
print()
print("train data:")
print(train_data)
```
